**Exercise 3(A) -->**
Calculate the probability of the counts of outcome *"2"* if the possible equally probable outcomes are all integer numbers between 1 and 10 an we performed *n* = 1000 trials.

```python
import numpy as np
#Initializing Trials to 1000. 
trials = 1000

#Setting counter to 0 and it will increase as we encounter 2.
counter = 0

for i in range(trials):
  # Random module will generate random integer from 1 - 10
  randomInt = np.random.randint(low=1,high=11,size=10)

  # Counter will be added of each value of 2
  counter += np.sum(randomInt == 2)*1

  #Probability 
  probability = counter/(10*trials)

  print("Probability of finding 2 is ",probability)
```

**Excercise 3(B) -->** Simulate n independent trials of the multinoulli (categorical distribution) compliant to the specification of (a). Plot the probability in (a) as a function of n independent trials (n=10-1000). Write your conclusions with respect to the behavior of the estimated probability as nn increases.

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating Empty Arrays to store Trails and Probabilities 
probabilities = []
trials = []

for i in range(10,1001):
  counter = 0

  for j in range(i):
    # Random module will generate random integer from 1 - 10
    randomInt = np.random.randint(low=1,high=11,size=10)

    # Counter will be added of each value of 2
    counter += np.sum(randomInt == 2)*1

  #Probability 
  probability = counter/(10*i)

  probabilities.append(probability)
  trials.append(i)

# Plotting of possibilities
plt.plot(trials,probabilities,color="green")
plt.xlabel("Trials")
plt.ylabel("Probabilities")
plt.show() 

# Result showcases as number of trials increases probability of getting 2 decreases

```

**Exercise 4** --> Generating Correlated Random Variables with Bivariate Gaussian (Normal) Distribution

```python
import numpy as np
import matplotlib.pyplot as plt


#Plot 1 with p = 0.00
p = 0.0
mean = (1,2)
cov = [[1,p],[p,1]]
x1, y1 = np.random.multivariate_normal(mean, cov, 10000).T


#Plot 2 with P = 0.20
p = 0.20
mean = (1,2)
cov = [[1,p],[p,1]]
x2, y2 = np.random.multivariate_normal(mean, cov, 10000).T



#Plot 3 with p = -0.80
p = -0.80
mean = (1,2)
cov = [[1,p],[p,1]]
x3, y3 = np.random.multivariate_normal(mean, cov, 10000).T



#Plot 4 with p = 0.99
p = 0.99
mean = (1,2)
cov = [[1,p],[p,1]]
x4, y4 = np.random.multivariate_normal(mean, cov, 10000).T

# Plotting 
figure2, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

ax1.plot(x1,y1,'x',color="red")

ax2.plot(x2,y2,'x',color="green")

ax3.plot(x3,y3,'x',color="blue")

ax4.plot(x4,y4,'x',color="purple")
plt.show()

# Results are for p = 0.00, p = 0.20 , p = -0.80 and p = 0.99 showing no correlation, some correlation, negative correlation and positive correlation. 



```
